## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix

plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Для определения показателей влияющих на цену сделала матрицу корреляции, из которой видно, что самая высокая зависимость от : общая жилая площадь; класс дома; кол-во ванных комнат;  площадь дома без подвала; средняя площадь дома. От этих показателей и будем анализировать цены на недвижимость.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Вывод: На графике мы видим, что в основном имеется тенденция роста цены от роста жилой площади дома.

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg');
```

Вывод: При росте количества ванных комнат есть тенденция к росту цены дома.

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

Вывод: При отсутствии подвала тенденция роста сохраняется как и при графике с общей жилой площадью.

```python
plt.figure(figsize=(12, 4))
sns.boxplot(y=df['price'],x=df['view'])
plt.title('Зависимость цены от качества вида из окна')
plt.xlabel('Шкала качества вида из окна')
plt.ylabel('Цена');
```

Ввод: График показывает, что чем лучше вид из окна, тем выше в стоимость домов.

```python
plt.figure(figsize=(12, 4))
sns.boxplot(y=df['price'],x=df['grade'])
plt.title('Зависимость цены от категории дома')
plt.xlabel('Категории домов')
plt.ylabel('Цена');
```

Вывод: Этот график очень хорошо показывает  как стоимость дома увеличивается от увеличения его категории.
